```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('weatherAUS.csv')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.isnull().sum().sum()
```

```python
categorical_col = []

for col in data.select_dtypes(include='object'):
    print(f'{col} : {data[col].unique()}')
    print()
    categorical_col.append(col)

data_cat = data[categorical_col]
```

```python
data[categorical_col].head()
```

```python
data[categorical_col].isnull().sum()
```

```python
encode = LabelEncoder()
data['RainTomorrow'] = encode.fit_transform(data['RainTomorrow'])
```

```python
numeric_col = []

for col in data.columns:
    if col != 'RainTomorrow' and (data[col].dtype == 'int64' or data[col].dtype == 'float64'):
        numeric_col.append(col)

data[numeric_col].corrwith(data['RainTomorrow'])
```

```python
encode = LabelEncoder()

for col in data_cat.columns:
    data[col] = encode.fit_transform(data[col])
```

```python
data
```

```python
data.drop(['Date','Evaporation','Sunshine','Cloud9am','Cloud3pm'],inplace=True,axis=1)
```

```python
data.head()
```

```python
y = data['RainTomorrow']
X = data.drop(['RainTomorrow'],axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression


# instantiate the model
logreg = LogisticRegression(random_state=0)

# fit the model
logreg.fit(X_train, y_train)
```

```python
predictions=logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, predictions)))
```

```python
from sklearn import svm
svm_model=svm.SVC()
svm_model.fit(X_train,y_train)
```

```python
prediction=svm_model.predict(X_test)
```
